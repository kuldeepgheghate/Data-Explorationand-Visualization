```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=pd.read_csv('/content/Students_Data.csv')
df.head()
```

```python
df.info()
```

```python
# Calculate the mean, median, and mode for 'math score'
mean = df['math score'].mean()
median = df['math score'].median()
mode = df['math score'].mode()[0]

# Print the results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
```

```python
# 1) box plot
plt.boxplot(df['reading score'])
plt.title('Box plot of reading score')
plt.show()
```

```python
# 2) histogram on math score
plt.hist(df['reading score'],color='red')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')
plt.title('Histogram of reading Score')
plt.show()
```

```python
#3) all histogram
df.hist()
plt.show()
```

```python
#4) count plot  on parental level of education
sns.countplot(x='parental level of education', data=df,color='pink')
plt.xticks(rotation=90)
plt.title('Parental Level of Education')
plt.show()
```

```python
#5) viline plot
sns.violinplot(df['writing score'],color="purple")
plt.title('Violin plot of writing score')
plt.ylabel('Writing Score')
plt.xlabel('density')
plt.show()
```

```python
#6) strip plot on reading score
sns.stripplot(x=df['reading score'],color='green')
plt.xlabel('Reading Score')
plt.ylabel('Distribution')
plt.title('Stripplot of reading score')
plt.show()
```

```python
# 7) swarmplot on reading score
sns.swarmplot(x=df['reading score'],color='black')
plt.xlabel('Reading Score')
plt.ylabel('Distribution')
plt.title('Swarmplot of reading score')
plt.show()
```

```python
#8) displot on math score
sns.displot(df['math score'],color='red')
plt.xlabel('Math Score')
plt.ylabel('Density')
plt.title('Displot of math score')
plt.show()
```

```python
#9) scatter plot on parental level of education and math score
sns.scatterplot(x='writing score', y='math score', data=df)
plt.xticks(rotation=90)
plt.xlabel('writing score')
plt.ylabel('Math Score')
plt.title('Boxplot of Parental Level of Education vs Math Score')
plt.show()
```

```python
#10) reg plot on writing score and reading score
sns.regplot(x='writing score', y='reading score', data=df)
plt.xlabel('writing Score')
plt.ylabel('Reading Score')
plt.title('Regression Plot of writing Score vs Reading Score')
plt.show()
```

```python
# Calculate the correlation coefficient between 'math score' and 'reading score'
print(df['math score'].corr(df['reading score']))

# Create a cross-tabulation table of 'math score' and 'reading score'
pd.crosstab(df['math score'],df['reading score'])
```
